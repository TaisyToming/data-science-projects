# Medical Charges Prediction 

This notebook explores the factors that influences medical charges. The dataset includes information on age, sex, BMI, number of children, smoking status, region and medical expenses.  
We begin by examining the overall structure and basic charecteristics of the dataset. We then investigate whether smoking is associated with higher medical costs and how BMI relates to medical charges. Finally, we fit a linear regression model to understand which variables contribute most to medical expenses and evaluate whether the assumptions of linear regression are met.

```python
# importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Data overview

```python
# Loading the dataset
df = pd.read_csv('medical-charges.csv')
df.head()
```

```python
# Checking the shape of the dataset
df.shape
```

```python
# Checking the colum data types
df.dtypes
```

```python
# Checking if there are any missing values
df.isna().sum()
```

```python
# Overview of the numerical columns
df.describe()
```

```python
# data distribution for categorical columns
df['sex'].value_counts() , df['smoker'].value_counts(), df['region'].value_counts()
```

## Exploratory Data Analysis

### Age distribution

```python

sns.histplot(x='age', bins = 20, data = df)
plt.show()
```

### BMI distribution

```python
sns.histplot(x='bmi',kde = True,  data = df)
plt.show()
```

### Distribution of Number of children 

```python
sns.countplot(data = df, x='children')
plt.show()
```

### Medical expenses by number of children

```python
sns.barplot(x='children', y = 'charges', data = df)
plt.show()
```

### Medical expenses by sex

```python
sns.barplot(x='sex', y = 'charges', data = df)
plt.show()
```

### Medical charges by Region

```python
sns.boxplot(x='region', y='charges', data = df)
plt.show()
```

### Medical charges by BMI

```python
sns.scatterplot(x='bmi', y = 'charges',alpha = 0.4,  data = df)
plt.show()
```

### Medical charges by smoker status and sex

```python
sns.boxplot(x='smoker', y = 'charges', hue = 'sex', data = df)
plt.show()
```

### Medical charges by smoker status

```python
colors = df['smoker'].map({"yes":1, "no":0})

sns.scatterplot(x='age', y = 'charges', data = df, hue = colors, alpha = 0.35)
plt.show()
```
